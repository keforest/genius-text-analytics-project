```python
import nltk
import re
import string

with open('C:/Users/chadc/Documents/Fall 2021 IAA/Text Analytics/Lyrics_2016_Clean') as f:
    s=f.read()

lines = s.count('\n')
lst = s.replace('\n', '. ')
sentence = nltk.tokenize.sent_tokenize(lst)

try:
    while True:
        sentence.remove('.')
except ValueError:
    pass

punc = re.compile( '[%s]' % re.escape( string.punctuation ) )
sent_vec = [ ]

for d in sentence:
    d = d.lower()
    d = punc.sub( '', d )
    sent_vec.append( nltk.sent_tokenize( d ) )

sent_list = [item for sublist in sent_vec for item in sublist]

sent_list
```
